```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [4,5,2,1,3,5]

plt.plot(x,y, color='cyan', lw=2, label='python')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line PLot')
plt.grid()
plt.legend(loc=9)
plt.show()
```

```python
plt.bar(x,y, color=['b','r','y','m','c','g'], edgecolor='k')
plt.xticks(x,["python",'java','C','js','Rust','Golang'])
plt.show()
```

```python
plt.scatter(x,y,c=x,cmap='cool')
plt.colorbar()
```

```python
x = [39,59,20,10,70]
plt.pie(x, labels=['java','C','Rust','Golang','Python'], autopct='%0.1f%%', startangle=90)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [3,5,6,2,1,4]

plt.step(x,y, marker='o')
plt.grid()
```

```python
x = [10,20,30,40,50,60,70,80,150]
x1 = [10,20,30,40,50,60,70,80,150]
x2 = [10,20,30,40,50,60,70,80,150]
y = [x,x1,x2]
plt.boxplot(y, patch_artist=True, showmeans=True)
```

```python
x = [1,2,3,4,5]
a1 = [3,4,5,2,1]
a2 = [4,2,1,4,5]
a3 = [5,3,2,1,4]

plt.stackplot(x,a1,a2,a3)
```

```python
plt.plot(x,y)
plt.fill_between(x,y)
plt.text(2,4, 'python')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.pie(x)

plt.subplot(2,2,3)
plt.bar(x,y)

plt.subplot(2,2,4)
plt.hist(x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = sns.load_dataset('tips')
```

```python
sns.lineplot(x='total_bill', y='tip', data=ds, hue='sex', palette='rainbow')
```

```python
sns.set_style('darkgrid')
sns.barplot(x='day', y='tip', data=ds, hue='sex', palette='rainbow', saturation=1)
```

```python
sns.histplot(ds['tip'], kde=True, color='green', log_scale=True)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=ds, hue='day', palette='rainbow', sizes=(56,34,56,76))
```
