```python
x = ['C++', "JavaScript", 'Python', "Java", "Rust"]
y = [75,80,90,70,65]
z = [65,70,80,55,90]

width = 0.2

p = np.arange(len(x))
p2 = [j+width for j in p]


plt.title("Programming Language Boom", fontsize=15)
plt.xlabel('Programming Languages', fontsize=15)
plt.ylabel('Popularity', fontsize=15)
plt.bar(p,y, width, color='c', align='center', edgecolor='k', linewidth=2, label='Popularity')
plt.bar(p2,z, width, color='y', align='center', edgecolor='k', linewidth=2, label='Popularity1')
plt.xticks(p+width/2, x, rotation=20)
plt.legend()
plt.show()
```

```python
ax = plt.axes(projection='3d')

x = np.arange(-5,5,0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap="Spectral")
ax.set_title("3D Plot")
ax.set_xlabel('test')
plt.show()
```
