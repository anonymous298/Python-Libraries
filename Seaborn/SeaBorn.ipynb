```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
varx = [1,2,3,4,5,6,7]
vary = [5,3,7,5,7,3,1]

df = pd.DataFrame({
    'varx' : varx,
    'vary' : vary
})

sns.lineplot(x='varx',y='vary', data=df)
plt.show()
```

```python
ds = sns.load_dataset('penguins').head(100)
ds
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=ds, hue='sex', style='sex', palette='viridis', markers=['o','v'], dashes=False, legend='auto')
plt.grid()
plt.title('Gender')
plt.show()
```

```python
data_set = sns.load_dataset('penguins')
orders = ['Biscoe', 'Torgersen', 'Dream']
sns.barplot(x='island' ,y='bill_length_mm', data=data_set, hue='sex', order=orders, hue_order=["Female", "Male"], saturation=1)
plt.title('BarPlot')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x='island', y='bill_depth_mm', data=data_set ,hue='sex', orient='v', palette='icefire', saturation=1, err_kws={'color':'c'})
plt.show()
```

```python
sns.set(style='darkgrid')
ds = sns.load_dataset('penguins')
sns.displot(ds['flipper_length_mm'], bins=[170, 180, 190, 200, 210, 220, 230 ,240], kde=True, rug=True, color='green')
plt.show()
```

```python
sns.scatterplot(x='bill_depth_mm', y='bill_length_mm', data=data_set, hue='sex', palette='icefire', style='sex',size='sex', sizes=(70,60),
               alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
var = np.linspace(0,10,20).reshape(4,5)

sns.heatmap(var)
plt.show()
```

```python
data = sns.load_dataset('anagrams')
x = data.drop(columns=["attnr"], axis=1).head(10)
x
```

```python
sns.heatmap(x, vmin=0, vmax=15, cmap='icefire', annot=True)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
sns.heatmap(var1, annot=True, cmap='icefire', vmax=15)
plt.show()
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],['b0' ,'b1' ,'b2' ,'b3' ,'b4']])
y = {'fontsize':20, 'color':'cyan'}
hm = sns.heatmap(var1, vmin=0, vmax=15, cmap='icefire', annot=ar, fmt='s',annot_kws=y, lw=2, linecolor='k', cbar=False, xticklabels=False, yticklabels=False)
hm.set(xlabel="Columns", ylabel='Rows')
plt.show()
```

```python
ds = sns.load_dataset('anagrams')
ds = ds.drop('attnr', axis=1).head(10)
sns.heatmap(ds, cmap='icefire', vmax=10, annot=True, lw=2, linecolor='k', annot_kws={'color': 'cyan'})
```
